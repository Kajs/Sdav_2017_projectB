```python
import csv

infile = 'Fire_Incidents_trimmed.csv'

firstLine = True

districtKey = 'Neighborhood  District'
numPersonnelKey = 'Suppression Personnel'
locationKey = 'Location'
timeKey = 'Alarm DtTm'
situationKey = 'Primary Situation'

situations = {}

columns = {}
with open(infile, 'rb') as input:
    reader = csv.reader(input, delimiter=',')
    for line in reader:
        if firstLine:
            firstLine = False
            i = 0
            for entry in line:
                columns[entry] = i
                i += 1
            continue
        situation = line[columns[situationKey]]
        if situation not in situations:
            situations[situation] = 0
        situations[situation] += 1

print timeKey + ': column ', columns[timeKey]
print numPersonnelKey + ': column ', columns[numPersonnelKey]
print situationKey + ': column ', columns[situationKey]
print districtKey + ': column ', columns[districtKey]
print locationKey + ': column ', columns[locationKey]
```

```python
situationsByDistrict = {}

firstLine = True
with open(infile, 'rb') as input:
    reader = csv.reader(input, delimiter=',')
    for line in reader:
        if firstLine:
            firstLine = False
            continue
            
        #insert district
        district = line[columns[districtKey]]
        if district == '':
            continue
            #district = 'No district specified'
        if (district not in situationsByDistrict):
            situationsByDistrict[district] = {} 
            
        #insert year
        timeString = line[columns[timeKey]];
        month = timeString[0:2]
        day = timeString[3:5]
        year = timeString[6:10]
        
        if year not in situationsByDistrict[district]:
            situationsByDistrict[district][year] = {}
            situationsByDistrict[district][year]['fire'] = 0
            situationsByDistrict[district][year]['other'] = 0
        
        #insert situation type, sorted by fire incident vs other
        situationString = line[columns[situationKey]]
        if 'fire' in situationString and 'no fire' not in situationString:
            situationsByDistrict[district][year]['fire'] += 1
        else:
            situationsByDistrict[district][year]['other'] += 1

with open('situationsByDistrict.csv', 'w+') as outfile:
    writer = csv.writer(outfile, delimiter=',')
    writer.writerow([districtKey, 'Year', 'Fire Incidents', 'Other Incidents'])
    for district in situationsByDistrict:
        for year in situationsByDistrict[district]:
            print district, year, 'fires: ' + str(situationsByDistrict[district][year]['fire']), 'other: ' + str(situationsByDistrict[district][year]['other'])
            writer.writerow([district, year, str(situationsByDistrict[district][year]['fire']), str(situationsByDistrict[district][year]['other'])])
```
